## Load Comparison Results

```python
comparison_path = MODELS_DIR / 'model_comparison.json'
comparison = json.loads(comparison_path.read_text())
comparison_df = pd.DataFrame(comparison)
comparison_df
```

```python
comparison_df.plot.bar(
    x='architecture',
    y=['best_val_accuracy', 'test_accuracy'],
    figsize=(9, 4),
    ylim=(0, 1),
    title='Model Accuracy Comparison'
);
```

```python
comparison_df.plot.bar(
    x='architecture',
    y='parameters',
    figsize=(9, 4),
    title='Trainable Parameter Count'
);
```

## Error Analysis

Inspect the best model's top mistakes and per-class accuracy.

```python
best_metrics_path = Path(comparison_df.iloc[0]['metrics'])
best_metrics = json.loads(best_metrics_path.read_text())
pd.DataFrame(best_metrics['test']['top_mistakes'])
```

```python
per_class_df = pd.DataFrame(best_metrics['test']['per_class'])
per_class_df.sort_values('accuracy').head(10)
```
